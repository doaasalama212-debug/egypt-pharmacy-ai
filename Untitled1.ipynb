```python
import os

os.environ["GROQ_API_KEY"] = None
```

```python
from crewai import LLM

llm = LLM(
    model="groq/llama-3.1-8b-instant",
    api_key=os.environ["GROQ_API_KEY"],
    temperature=0.3
)
```

```python
import requests
from bs4 import BeautifulSoup
from crewai.tools import tool


@tool("pharmacy_scraper")
def pharmacy_scraper(query: str) -> str:
    """
    Scrapes pharmacy competitor websites and returns ONLY small titles.
    """

    websites = [
        "https://www.elezabypharmacy.com",
        "https://seifpharmacies.com",
        "https://roshdypharmacy.com"
    ]

    headers = {"User-Agent": "Mozilla/5.0"}

    results = []

    for url in websites:
        try:
            r = requests.get(url, headers=headers, timeout=10)
            soup = BeautifulSoup(r.text, "lxml")

            title = soup.title.text.strip() if soup.title else "No title"

            # IMPORTANT: keep output SMALL
            results.append(f"{url} → {title}")

        except Exception as e:
            results.append(f"{url} → error")

    return "\n".join(results)
```

```python
from crewai import Agent

market_researcher = Agent(
    role="Pharmacy Market Researcher",
    goal="Analyze pharmacy competitors in Egypt",
    backstory="Expert in Egyptian healthcare retail market",
    tools=[pharmacy_scraper],
    llm=llm,
    verbose=True
)
```

```python
from crewai import Task

research_task = Task(
    description="Analyze pharmacy competitors using tool output. Focus on pricing, service, and customer experience.",
    expected_output="Short competitor analysis report",
    agent=market_researcher
)
```

```python
from crewai import Crew, Process

crew = Crew(
    agents=[market_researcher],
    tasks=[research_task],
    process=Process.sequential,
    verbose=True,
    max_rpm=3  # important to avoid Groq limit
)

result = crew.kickoff()

print("\n===== FINAL RESULT =====\n")
print(result)
```

```python
research_task = Task(
    description="""
Analyze pharmacy competitors using the scraping tool output.

Return the analysis in this structure:

1. Competitor Overview
2. Pricing Strategy
3. Customer Experience
4. Strengths
5. Weaknesses
6. Opportunities
7. Threats (SWOT)

Keep it concise and structured.
""",
    expected_output="Structured SWOT analysis of Egyptian pharmacies",
    agent=market_researcher
)
```

```python
strategist = Agent(
    role="Pharmacy Marketing Strategist",
    goal="Turn competitor analysis into actionable business strategy",
    backstory="Expert in launching pharmacies in Egyptian market",
    llm=llm,
    verbose=True
)
```

```python
strategy_task = Task(
    description="""
Based on the SWOT analysis, create a launch strategy for a new pharmacy in Egypt.

Include:
- Brand positioning
- Pricing strategy
- Marketing channels
- Customer acquisition plan
- Differentiation strategy
""",
    expected_output="Full pharmacy business strategy",
    agent=strategist,
    context=[research_task]
)
```

```python
crew = Crew(
    agents=[market_researcher, strategist],
    tasks=[research_task, strategy_task],
    process=Process.sequential,
    verbose=True,
    max_rpm=3
)

result = crew.kickoff()

print(result)
```

```python
def format_output(text):
    print("\n" + "="*60)
    print("PHARMACY MARKET INTELLIGENCE REPORT")
    print("="*60 + "\n")
    print(text)
```

```python
format_output(result)
```

```python
with open("pharmacy_report.txt", "w", encoding="utf-8") as f:
    f.write(str(result))

print("Report saved successfully!")
```
